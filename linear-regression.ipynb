```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/salary-data/Salary Data.csv')
```

```python
df.head()
```

```python
x = df.drop(columns=['Salary'])
x.head()
```

```python
y = df.drop(columns=['Experience Years'])
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[1.1]])
```

```python
reg.predict([[39343]])
```

```python
df['predicted_y']=reg.predict(x)
df.head()
```

```python
df['loss'] = df['Salary'] - df['predicted_y']
df.head()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
mse = mean_squared_error(df['Salary'] , df['predicted_y'])
mse
```

```python
mae = mean_absolute_error(df['Salary'] , df['predicted_y'])
mae
```
